```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("hepatitis_EDA/hepatitis_csv.csv")
```

```python
data = data.dropna()
```

```python
data = data.reset_index(drop = True)
```

```python
data
```

```python
data.groupby('age').mean()['sgot'].sort_values(ascending=False)
```

```python
x = data["age"]
y = data["sgot"]
plt.bar(x,y, color = "green")
plt.xlabel("age")
plt.ylabel("Average sgot level")
plt.title("SGOT vs AGE graph")
plt.show()
```

```python
print("Safe sgot levels")
data[data['sgot']<50].sort_values('sgot',ascending=False)
```
